```python
import matplotlib.pyplot as plt
import numpy as np

a=np.linspace(0,10,100)
b=a**4
```

```python
x=np.arange(1,10)
y=2*x
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,10),dpi=100)
# axes[0][0].plot(x,y)
# axes[1][1].plot(a,b)5

# for ax in axes:
    # ax.plot(x,y)
# axes[0].plot(x,y)
# axes[1].plot(a,b)
# axes[2].plot(b,a,'b--')
# plt.tight_layout() # to prevent overlapping

axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(b,a)
axes[1][1].plot(y,x,'r--')
axes[0][0].title.set_text('First Plot')
axes[0][0].set_xlabel('X axis')
axes[0][0].set_ylabel('Y axis')
fig.subplots_adjust(wspace=0.5,hspace=0.5)
fig.suptitle('My Subplots',fontsize=17)
fig.width_ratios=[1,2]
plt.tight_layout()

```
