```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
fig = plt.figure(dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x,label='x line',color='green')
axes.plot(x, x**2,label='x squared',color='red')  
axes.legend(loc=(0.9,0.8))
```
