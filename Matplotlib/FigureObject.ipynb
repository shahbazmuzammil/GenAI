```python

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
axes= fig.add_axes([0,0,1,1])
x=np.random.randint(1,20,10)
y=x**2
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Main Plot')
axes.grid(True)
axes2=fig.add_axes([0.1,0.7,0.10,0.10])
axes2.plot(x,y,)
axes2.set_title('Inset Plot')   
axes2.grid(True)
axes2.set_ylabel('Y axis')
axes2.set_xlabel('X axis')
axes2.set_ylim(0,50)
axes2.set_xlim(0,5)

plt.show()
```
