List Comprehension
* Basic Syntax [expression for item in iterable]
* With Condition logic [expression for item in iterable if condition]
* Nested List Comprehension [expression for item1 iterable 1 for item 2 iterable2]
* [ RESULT_EXPRESSION for ITEM in ITERABLE ]

```python
### Basic list comprehension 
square = [num**2 for num in range(10)]
print(square)
```

```python
lst = []
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
odd= [i for i in range(10) if i%2==0]
print(odd)
```

```python
number = [1,2,3,4,5]
alpha = ["a","b","c","d"]
pair = [[i,j] for i in number for j in alpha]
print(pair)
```

List Comprehesion with function call

```python
words = ["hellow","world","python","list"]
length = [len(word) for word in words]
print(length)
# Breakdown of the Order:
# len(word) (The Transformation): This tells Python, "For every item I find, calculate its length and add that result to my new list."
# for word (The Variable): This defines a temporary name (word) for the current item being processed.
# in words (The Source): This tells Python where to get the data from.
```

list comprehensions are a powerfull and concise way to create lists in Python they are syntactically compact and can replace more verbose constructs understanding the syntax of list comprehension will help you write cleaner and more efficeint python code
