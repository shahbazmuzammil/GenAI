*Dictionaries

Dictionaries are unordered collection of items. They store data in key-value pairs.Key =s must be unique and immutable (e.g string,number,or tuples), while values can be of any type.

```python
student = {"name":"shahbaz","age":28,"grade":"A"}
print(student)
print(student["name"])
print(student.get("age","available"))
```

```python
#Modifying Dictionary Elements
#Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student["age"]=29 #updating the value for key
print(student)
student["address"]="India" # added a new key and value
```

```python
# del student["age"] # delete a key and value pair
# print(student)
student["age"]=28
student["grade"]="A"
```

```python
# Dictionary methods
keys=student.keys()
print(keys)
values=student.values()
print(values)
items=student.items()
print(items)
```

```python
# shallow copy
student_shallow_copy=student
print(student_shallow_copy)
```

```python
# iterating over the dictionaries
# you can use loops to iterates over dictionaries, key, values or items
for value in student.values():
    print(value)
    
for key,value in student.items():
    print(f'{key}:{value}')


```

```python
##Nested Dictionaries

students={
    "student1":{"name":"zainab","age":3},
    "student2":{"name":"shireen","age":1}
}
print(students["student1"]["name"])
print(students["student2"]["name"])
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
# Dictionary comprehesion

square = {x:x**2 for x in range(5)}
print(square)
```

```python
# Condition dictionary comprehension

even = {x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
numbers =[1,2,2,3,3,4,4,5,5,5,6,6,6]
frequency={}

for number in numbers :
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
# Merge dictionaries into one   
dict ={"a":1,"b":2}
dict2 ={"b":3,"c":4}
meged={**dict,**dict2}
print(meged)

```
