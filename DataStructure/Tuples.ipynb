```python
# Packing and Unpacking the Tuple
# whenever the number or string in comma saperated value it packed in a tuple
# Packing
packed_tuple = 1, "hello",3.14
print(packed_tuple)
```

```python
# Unpacking a tuple
# destructuring
a,b,c=packed_tuple
print(b)
```

```python
# unpacking the tuple with * 
numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```
