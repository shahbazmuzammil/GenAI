Lambda Function in python 
Lambda functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only expression, they are commonly used for short operations or as arguments to higher-order functions like map(), filter(), and reduce().

```python
lambda arguments: expression
```

```python
addition=lambda a,b : a+b
type(addition)
print(addition(9,9))
```

```python
def even(num):
    if num%2==0:
        return True
print(even(4))
```

```python
even=lambda num:num%2==0
print(even(5))
```

```python
def addition(a,b,c):
    return a+b+c
print(addition(2,3,4))
add=lambda a,b,c : a+b+c
print(add(2,3,4))
```

```python
numbers = [1, 2, 3, 4, 5]
def square(x):
    return x**2

print(square(2))
squared_numbers = list(map(square, numbers))
squared_numbers=list(map(lambda x:x**2,numbers))
print(squared_numbers)
```
