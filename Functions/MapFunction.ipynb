the Map() funtion applies a given function to all items in an input list (or any iterable) and returns a map object (which is an iterator).
this is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x
square(10)
```

```python
numbers = [1, 2, 3, 4, 5]
# map takes two inputs: a function and an iterable (like a list)
add=list(map(square,numbers))
print(add)
list(map(lambda x:x*x,numbers))
```

```python
# two function
number1=[1,2,3]
number2=[4,5,6]
list(map(lambda x,y:x+y,number1,number2))
```

```python
# map() tp convert a list of string to integer
# use map to convert string into the integer basically a type casting

str=["1","2","3","4"]

list(map(int,str))
```

```python

def get_name(name):
    return name["name"]

people=[
    {"name":"Alice","age":30},
    {"name":"Bob","age":25},
    {"name":"Charlie","age":35}
]
print(list(map(get_name,people)))
    
```
