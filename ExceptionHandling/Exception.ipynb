Uderstanding Exceptions

Exception handling in python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exception, including how to use try,
except else and finally blocks

```python
# Zero ZeroDivisionError
try:
    result=9/0
    a=b
except ZeroDivisionError as x :
    print("Error:", x)
except Exception as e:
    print("General Error:", e)
```

```python
try:
    num=int(input("Enter a number: "))
    result=10/num
    print("You entered:", num)
except ValueError as ve :
    print("Invalid input! Please enter a valid integer.", ve)
except ZeroDivisionError as zde :
    print("Division by zero is not allowed.", zde)
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
try:
    num=int(input("Enter a number: "))
    result=10/num
    print("You entered:", num)
except ValueError as ve:
    print("Invalid input! Please enter a valid integer.", ve)
except ZeroDivisionError as zde:
    print("Division by zero is not allowed.", zde)
except Exception as e:
    print("An unexpected error occurred:", e)
else:
    print("The result is:", result)
finally:
    print("Execution completed.")
```
