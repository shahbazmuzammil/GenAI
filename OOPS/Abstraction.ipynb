```python
from abc import ABC, abstractmethod

class vehicle(ABC):
    
    def drive(self):
        print("driving")
    @abstractmethod
    def start_engine(self):
        pass
class Car(vehicle):
    def start_engine(self):
        print("car engine started")
    
def operate_vehicle(v: vehicle):
    v.start_engine()
    v.drive()
    
my_car = Car()
operate_vehicle(my_car)
```
