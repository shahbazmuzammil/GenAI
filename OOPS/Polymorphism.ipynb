```python
#Method Polymorphism in Python

class Car ():
    def start_Engine(self):
        print("Engine Started")
        
class Truck(Car):
    def start_Engine(self):
        print("Truck Engine Started with a roar!")

class Bike(Car):
    def start_Engine(self):
        print("Bike Engine Started with a vroom!")
        
def vehicle(car):
    print(car.start_Engine())
    
# Creating objects of each class
my_car = Car()
my_truck = Truck()
my_bike = Bike()
# machine_start()
print(Car().start_Engine())
print(Truck().start_Engine())
print(Bike().start_Engine())
vehicle(Car())
```
