```python
# ...existing code...
class Car:
    def __init__(self, speed, color, engineType):
        self.speed = speed
        self.color = color
        self.engineType = engineType

    def car_Details(self):
        return f"Speed: {self.speed}, Color: {self.color}, Engine Type: {self.engineType}"

car1 = Car(200, "Red", "V8")
print(car1.car_Details())
# ...existing code...
```

```python
class Tesla(Car):
    def __init__ (self,speed,color,engineType,selfdriving):
        self.selfdriving = selfdriving
        super().__init__(speed,color,engineType)
    def car_Details(self):
        car=super().car_Details()
        return f"{car}, Self Driving: {self.selfdriving}"   
tesla1=Tesla(200,"yellow","V8","Yes")
print(tesla1.car_Details())
        
```

```python
#Multiple Inheritance
#when a class  inheritance from more than one base class
class Base1:
    def __init__(self,name):
        self.name=name
    def greet1(self):
        return f"Hello from Base1, {self.name}!"
    
class Base2:
    def __init__(self, age):
        self.age = age

    def greet2(self):
        return f"Hello from Base2, {self.age} years old!"
    
class Derived(Base1, Base2):
    def __init__(self, name, age):
        Base1.__init__(self, name)
        Base2.__init__(self, age)

    def greet(self):
        return f"{self.greet1()} {self.greet2()}"
```

```python
class Animal ():
    def __init__(self,species):
        self.species=species
    def animal_details(self):
        return f"Species: {self.species}"
animal=Animal("Cat")
print(animal.animal_details())
class Pet ():
    def __init__(self,owner):
        self.owner=owner
    def pet_details(self):
        return f"Owner: {self.owner}"  
pet =Pet("Alice")
print(pet.pet_details())

class Cat(Animal,Pet):
    def __init__(self,species,owner,breed):
        Animal.__init__(self,species)
        Pet.__init__(self,owner)
        self.breed=breed
    def cat_details(self):
        animal_info=self.animal_details()
        pet_info=self.pet_details()
        return f"{animal_info}, {pet_info}, Breed: {self.breed}"
cat1=Cat("Cat","Alice","Siamese")
print(cat1.cat_details())
    
```
