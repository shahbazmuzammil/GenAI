```python
class BankAccount() :
    def __init__(self,name,balance=0):
        self.name=name
        self.balance=balance
        
    def withdraw(self,amount):
        if amount>self.balance:
            print(f"withdrawn amount is very low:{amount}")
        else :
            print("money withdrawn")
    def deposit(self,amount):
        self.balance += amount
        print(f"{self.balance}amount is deposite")
    def get_balance(self):
        return self.balance
account = BankAccount('Aiman',100000)
print(account.balance,account.name,"sdfw")
```
