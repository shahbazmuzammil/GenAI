```python
with open("kaboom.csv","w")as f:
    f.write("name,age,city\n")
    f.write("Alice,30,New York\n")
    f.write("Bob,25,Los Angeles\n")
with open("kaboom.csv",mode="r")as f:
    print(f.read())
    
```

```python
#writing file line by line
lines=['First line\n','Second line\n','Third line\n']
quickfile='quickfile.txt'
with open(quickfile,'w')as f:
    f.writelines(lines)
```

```python
#Binary Files
data=b'\x00\xFF\x7A\x34'
with open("binaryfile.bin","wb")as f:
    f.write(data)
```

```python
with open("source.txt","r")as f:
    content=f.read()
    print(content)
    
with open("quickfile.txt","w") as f:
    f.write(content)
    print("File overwritten.")
```

```python
#writing and then reading the file
with open("kaboom.txt","w+") as file:
    file.write("Hello, World!")
    file.write("\nWelcome to file operations in Python.")
    file.seek(0)  # Move the cursor to the beginning of the file
    content = file.read()
    print(content)
```
