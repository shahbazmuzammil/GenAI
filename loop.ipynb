```python
# looop control statement
# break
# the break statement
for i in range(10):
        if i==5:
            break
        print(i)
```

```python
# continue
# the continue statnent skips the current iterartion and continues with the next
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# Nested Loop
# Loop inside a loop
# outer loop slow and inner loop is fast like if i is 0 then it enter inside the inner loop inner loop complete it cycle (o,1,2) the  it get out and outer loop complete the cycle and get complete the loop slow and fast nested loop
for i in range(3):
    for j in range(3):
        print(f"i:{i} and j:{j}")
```

```python
# calculate the sum of first N natural number using a while  and for loop

# while loop

n=10
sum = 0
count = 1

while count<=n:
    sum = sum+count
    count=count+1
    
    print ("sum of first 10 natural number",sum)
```

```python
# prime number

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
