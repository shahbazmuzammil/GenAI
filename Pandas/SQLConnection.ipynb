```python

import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
temp_db=create_engine('sqlite:///test.db')

```

```python
df=pd.DataFrame(data=np.random.randint(low=0,high=101,size=(4,4)))
df.columns=['a','b','c','d']
df
```

```python
# pd.read_sql(sql='new_table',con=temp_db)
df.to_sql(name='new_table',con=temp_db,index=False)


```

```python
result=pd.read_sql_query('SELECT a,c FROM new_table',con=temp_db)
result
```
