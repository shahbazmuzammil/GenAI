```python
import pandas as pd

# Create a sample product dataset
products_data = {
    'Product ID': [101, 102, 103, 104, 105],
    'Product Name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Accessories'],
    'Price': [999.99, 29.99, 79.99, 299.99, 149.99],
    'Quantity': [15, 50, 35, 20, 45],
    'Stock Status': ['In Stock', 'In Stock', 'Low Stock', 'In Stock', 'In Stock']
}

# Create DataFrame
df = pd.DataFrame(products_data)

# Write to Excel file
df.to_excel('products.xlsx', index=False, sheet_name='Products')

print("Excel file 'products.xlsx' created successfully!")
df
```
