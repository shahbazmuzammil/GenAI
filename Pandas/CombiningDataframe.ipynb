```python
# Inner Merge
import pandas as pd
import numpy as np 

registration = {
    'name': ['Alice', 'Bob', 'Charlie', 'yolanda'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 'david@email.com'],
    'registration_date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04']
}

login = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'last_login': ['2025-01-20', '2025-01-22', '2025-01-18', '2025-01-25'],
    'login_count': [45, 32, 28, 51]
}

register= pd.DataFrame(registration)
login=pd.DataFrame(login)

merging=pd.merge(register,login,how='outer',on='name')
merging
# index merging
register
login
```
